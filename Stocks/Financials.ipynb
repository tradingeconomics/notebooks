## Example 1 - Current Companies Assets/Liabilities Comparison

### Importing data

```python
raw_data_microsoft = te.getFinancialsData('msft:us', 'df')
raw_data_apple = te.getFinancialsData('aapl:us', 'df')
raw_data_amazon = te.getFinancialsData('amzn:us', 'df')
raw_data_google = te.getFinancialsData('goog:us', 'df')
raw_data_facebook = te.getFinancialsData('aapl:us', 'df')
```

### Manipulating data

```python
assets_microsoft = raw_data_microsoft[ 'last'][3:4]
assets_apple = raw_data_apple[ 'last'][3:4]
assets_amazon = raw_data_amazon[ 'last'][3:4]
assets_google = raw_data_google[ 'last'][3:4]
assets_facebook = raw_data_facebook['last'][3:4]
assets_all = pd.concat([assets_microsoft, assets_apple, assets_amazon, assets_google, assets_facebook]  )

liabilities_microsoft = raw_data_microsoft[ 'last'][4:5]
liabilities_apple = raw_data_apple[ 'last'][4:5]
liabilities_amazon = raw_data_amazon[ 'last'][4:5]
liabilities_google = raw_data_google[ 'last'][4:5]
liabilities_facebook = raw_data_facebook['last'][4:5]
liabilitiess_all = pd.concat([liabilities_microsoft, liabilities_apple, liabilities_amazon, liabilities_google, liabilities_facebook]  )
```

### Plotting

```python
labels = ['Microsoft', 'Apple', 'Amazon', 'Google', 'Facebook']
assets = assets_all.values 
liabilitiess = liabilitiess_all.values 

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10,10))
rects1 = ax.bar(x - width/2, assets, width, label='Assets')
rects2 = ax.bar(x + width/2, liabilitiess, width, label='Liabilities')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title("Current Companies Assets / Liabilities" )
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.set_ylim([50000, 180000])
ax.yaxis.set_major_formatter('${x:1.2f}')

fig.tight_layout()

plt.show()
```

## Example 2 - Companies Capital's Ratio

### capital ratio = Current assets รท current liabilities



```python
current_assets_microsoft = raw_data_microsoft['last'][3]
current_liabilities_microsoft = raw_data_microsoft['last'][4]
capital_ratio_microsoft = current_assets_microsoft / current_liabilities_microsoft

current_assets_google = raw_data_google['last'][3]
current_liabilities_google = raw_data_google['last'][4]
capital_ratio_google = current_assets_google / current_liabilities_google

current_assets_amazon = raw_data_amazon['last'][3]
current_liabilities_amazon = raw_data_amazon['last'][4]
capital_ratio_amazon = current_assets_amazon / current_liabilities_amazon

current_assets_apple = raw_data_apple['last'][3]
current_liabilities_apple = raw_data_apple['last'][4]
capital_ratio_apple = current_assets_apple / current_liabilities_apple

current_assets_facebook = raw_data_facebook['last'][3]
current_liabilities_facebook = raw_data_facebook['last'][4]
capital_ratio_facebook = current_assets_facebook / current_liabilities_facebook
```

### Plotting

```python
fig, ax = plt.subplots(figsize=(15,5))


ax.bar('Microsoft', capital_ratio_microsoft, label= 'Microsoft')
ax.bar('Amazon', capital_ratio_amazon, label='Amazon')
ax.bar('Facebook', capital_ratio_facebook, label = 'Facebook')
ax.bar('Google', capital_ratio_google, label='Google')
ax.bar('Apple', capital_ratio_apple, label ='Apple')
plt.style.use('bmh')
ax.set(ylim=[0.5, 4], title='Working Capital Ratio')

ax.legend()

```
