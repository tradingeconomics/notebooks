Importing dependencies

```python
import tradingeconomics as te
import matplotlib.pyplot as plt
```

Importing data

Plotting

```python
plt.plot(df)
plt.title("United states - GDP")
plt.grid(True)
plt.ylabel("$B")
plt.xlabel("Date")
plt.show()
```
