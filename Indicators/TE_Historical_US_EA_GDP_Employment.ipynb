Impact of the 2020 recession on GDP and Employment in the US and EA

```python
import tradingeconomics as te
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
te.login('jupyter:jupyter'); # Signup for a free API key at https://developer.tradingeconomics.com/
```

```python
data_gdp = te.getHistoricalData(country='United States', indicator='GDP Constant Prices', initDate='2019-01-01')
data_employment = te.getHistoricalData(country='United States', indicator='Employed Persons', initDate='2019-01-01')

data_gdp_referece_point = data_gdp[0][0]
data_employment_referece_point = data_employment[0][0]

data_gdp_index = data_gdp.div(data_gdp_referece_point) * 100
data_employment_index = data_employment.div(data_employment_referece_point) * 100

# print(data_gdp_index)
# print(data_employment_index)
```

```python
plt.plot(data_gdp_index, label='GDP Constant Prices')
plt.plot(data_employment_index, label='Employed Persons')
plt.title("United States - GDP and Employment (Q1 2019 = 100)")
plt.grid(True)
plt.ylabel("Index")
plt.xlabel("Date")
plt.legend()
plt.show()
```

```python
data_gdp_ea = te.getHistoricalData(country='Euro Area', indicator='GDP Constant Prices', initDate='2019-01-01')
data_employment_ea = te.getHistoricalData(country='Euro Area', indicator='Employed Persons', initDate='2019-01-01')

data_gdp_referece_point_ea = data_gdp_ea[0][0]
data_employment_referece_point_ea = data_employment_ea[0][0]

data_gdp_index_ea = data_gdp_ea.div(data_gdp_referece_point_ea) * 100
data_employment_index_ea = data_employment_ea.div(data_employment_referece_point_ea) * 100

# print(data_gdp_index_ea)
# print(data_employment_index_ea)
```

```python
plt.plot(data_gdp_index_ea, label='GDP Constant Prices')
plt.plot(data_employment_index_ea, label='Employed Persons')
plt.title("Euro Area - GDP and Employment (Q1 2019 = 100)")
plt.grid(True)
plt.ylabel("Index")
plt.xlabel("Date")
plt.legend()
plt.show()
```

```python
plt.plot(data_gdp_index, label='United States')
plt.plot(data_gdp_index_ea, label='Euro Area')
plt.title("GDP (Q1 2019 = 100)")
plt.grid(True)
plt.ylabel("Index")
plt.xlabel("Date")
plt.legend()
plt.show()
```

```python
plt.plot(data_employment_index, label='United States')
plt.plot(data_employment_index_ea, label='Euro Area')
plt.title("Employment (Q1 2019 = 100)")
plt.grid(True)
plt.ylabel("Index")
plt.xlabel("Date")
plt.legend()
plt.show()
```

In the US and EA is registred an inverted effect on GDP and Employment. While in the US the employment falls a lot more than GDP the opposite happens in the EA. It's also possible to compare a more resiliant GDP in the US vs EA but a stronger employment in the EA.
