US Money Supply and Inflation Rate

```python
import tradingeconomics as te
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
te.login(); # Signup for a free API key at https://developer.tradingeconomics.com/
```

```python
data_money0 = te.getHistoricalData(country='United States', indicator='Money Supply M0', initDate='2019-01-01', output_type='df')
data_money1 = te.getHistoricalData(country='United States', indicator='Money Supply M1', initDate='2019-01-01', output_type='df')
data_money2 = te.getHistoricalData(country='United States', indicator='Money Supply M2', initDate='2019-01-01', output_type='df')

data_money_referece_point0 = data_money0.iloc[0, 3]
data_money_referece_point1 = data_money1.iloc[0, 3]
data_money_referece_point2 = data_money2.iloc[0, 3]

data_money_index0 = data_money0['Value'].div(-data_money_referece_point0) * 100
data_money_index1 = data_money1['Value'].div(-data_money_referece_point1) * 100
data_money_index2 = data_money2['Value'].div(-data_money_referece_point2) * 100
```

```python
plt.plot(data_money_index0, label='Money Supply M0')
plt.plot(data_money_index1, label='Money Supply M1')
plt.plot(data_money_index2, label='Money Supply M2')
plt.title("United States - Money Supply and Consumer Prices (jan 2019 = 100)")
plt.grid(True)
plt.ylabel("Index")
plt.xlabel("Date")
plt.legend()
plt.show()
```

In this first graph we see huge increases on the United States economy money supply. M0 is close to 200% of the 2019/01 value, M1 is at over 500% and M2 at 150%. But in the next graph we see an inflation of just 9% since january 2019.

```python
data_inflation = te.getHistoricalData(country='United States', indicator='Consumer Price Index CPI', initDate='2019-01-01')
data_inflation_referece_point = data_inflation[0][0]
data_inflation_index = data_inflation.div(data_inflation_referece_point) * 100
```

```python
plt.plot(data_inflation_index, label='CPI')
plt.title("United States - Consumer Prices (jan 2019 = 100)")
plt.grid(True)
plt.ylabel("Index")
plt.xlabel("Date")
plt.legend()
plt.show()
```

After the money supply had increased drastically following the FEDs stimulus, we see a pickup on the inflation, but these increases are still shy compared to the dimension of the stimulus. The question remains, are the prices starting to ease or will they rise closer to the stimulus values?
