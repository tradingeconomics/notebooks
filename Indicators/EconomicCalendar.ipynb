Data manipulation

```python
df['Hour'] = df['Date'].apply(lambda x: x[11:])
df['Date'] = df['Date'].apply(lambda x: x[:10])

df = df[['Date', 'Hour', 'Country', 'Category', 'Event', 'Source', 'Actual', 'Previous', 'Forecast', ]]

df.sort_values(['Date', 'Hour'], inplace=True)

df.head(30)
```
