# Percentage Changes of Crypto Market

```python
import tradingeconomics as te
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
te.login()
```

```python
data = te.getMarketsData(marketsField = 'crypto', output_type = 'df')
```

```python
data
```

```python
df=data[['Name', 'DailyPercentualChange', 'WeeklyPercentualChange', 'MonthlyPercentualChange', 'YearlyPercentualChange', 'YTDPercentualChange']]
df
```

```python
df = df.set_index('Name')
```

```python
plt.figure(figsize = (16,8))
plt.title('Crypto % Change', size=20)
ax = sns.heatmap(df, annot=True, linewidth=1, cmap='RdYlGn', cbar_kws={'label': '% Change of Crypto Market'});

ax.figure.axes[-1].yaxis.label.set_size(20)
```
